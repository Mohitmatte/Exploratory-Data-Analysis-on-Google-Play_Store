```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('googleplaystore.csv')
df['Reviews']
df[~df.Reviews.str.isnumeric()]
df_copy=df.copy()
df_copy
df_copy.reset_index(drop=True,inplace=True)
df_copy
df_copy[~df_copy.Reviews.str.isnumeric()]
def size_process(item):
    if str(item)[-1]=='M':
        res=float(str(item).replace('M',''))
        res=res*1024
        return res
    elif str(item)[-1]=='k':
        res=float(str(item).replace('k',''))
        return res
    else:
        return str(np.NaN)
df_copy['Size']=df_copy['Size'].apply(size_process)
df_copy.Size.dtype
df_copy.Size=df_copy['Size'].astype('float')
df_copy['Installs']=df_copy['Installs'].str.replace("+","").str.replace(",","")
df_copy['Installs']
df_copy.Installs=df_copy['Installs'].astype(int)
df_copy["Price"].unique()
char_to_remove=["+", "," ,"$"]
cols_to_remove=["Price"]

for char in char_to_remove:
    for col in cols_to_remove:
        df_copy[col] = df_copy[col].str.replace(char,"")
df_copy['Price']=df_copy['Price'].astype(float)
df_copy['day']= df_copy['Last Updated'].dt.day
df_copy['month']= df_copy['Last Updated'].dt.month
df_copy['year']= df_copy['Last Updated'].dt.year


#EDA
cfeatures=[feature for feature in df_copy.columns if df_copy[feature].dtype=='O']
nfeatures=cfeatures=[feature for feature in df_copy.columns if df_copy[feature].dtype!='O']
sns.countplot(x=df_copy["Content Rating"])
sns.displot(df_copy["Price"],kde=True)
sns.displot(df_copy["Rating"],kde=True)
plt.figure(figsize=(20,15))
category=["Type","Content Rating"]


for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df_copy[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
df_copy["Category"].value_counts().plot.pie(y=df["Category"],figsize=(12,12),autopct='%1.1f%%')    
```
